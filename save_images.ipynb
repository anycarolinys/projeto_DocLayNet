# **Salvando Imagens do DocLayNet Localmente**

## **Criando os diretórios**

```python
from datasets import load_dataset
import numpy as np
import os
from PIL import Image
```

```python
%%time
dataset = load_dataset("pierreguillou/DocLayNet-small", trust_remote_code=True)
```

```python
import os
image_folder = './images'
if not os.path.exists(image_folder):
    os.mkdir('./images')

categories = np.array(dataset['train']['doc_category'])
print(np.unique(categories))

for category in categories:
    category_folder = os.path.join(image_folder, category)
    if not os.path.exists(category_folder):
        os.mkdir(category_folder)
        print("Diretório criado:", category_folder)
    else:
        print("O diretório", category_folder, "já existe.")
```

```python
# Testando a obtenção dos nomes originais dos arquivos
print(dataset['train'][0]['original_filename'])
print(dataset['train'][0]['original_filename'][:-4])
print(dataset['train'][0]['original_filename'].replace('.pdf','.png'))
```

```python
def save_image(dataset_row):
    output_path = './images/' + dataset_row['doc_category'] + '/' + dataset_row['original_filename'].lower().replace('.pdf','.png')
    # Cria os diretórios necessários se não existirem
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    # Salva a imagem
    dataset_row['image'].save(output_path)
    print(f"Imagem salva em: {output_path}")
```

```python
dataset['train'].map(save_image)
```

```python
def processar_imagens_em_pastas(pasta_principal):
    imagens_processadas = []

    for root, dirs, files in os.walk(pasta_principal):
        for nome_da_pasta in dirs:
            caminho_completo_pasta = os.path.join(root, nome_da_pasta)
            
            for filename in os.listdir(caminho_completo_pasta):
                if filename.endswith('.png'):
                    caminho_completo_arquivo = os.path.join(caminho_completo_pasta, filename)
                    
                    with Image.open(caminho_completo_arquivo) as img:
                        processed_image = process_image(img)

                        imagens_processadas.append(processed_image)

    imagens_processadas_array = np.array(imagens_processadas)

    return imagens_processadas_array
```

```python
folder_path = './images'
imagens_processadas_array = processar_imagens_em_pastas(folder_path)
```
